# 2024-08-05 と 2025-04-07

Reference: https://ja.wikipedia.org/wiki/2024%E5%B9%B4%E3%81%AE%E6%A0%AA%E4%BE%A1%E5%A4%A7%E6%9A%B4%E8%90%BD

2024年8月2日、アメリカ合衆国労働省の雇用率の大幅な低下を示す雇用統計を受けて金融市場は下落し、景気後退の懸念が高まりました。

この株価大暴落は、以下の影響を受けました。

- 日本銀行の利上げ決定に伴う日本株の下落
- 企業利益の減少を投資家が懸念したことによる日本円の円高の進行
- アメリカ合衆国のテクノロジー株が過大評価されているとの認識

```python
import datetime

import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import mplfinance as mpf
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
def chart_candle(df):
    FONT_PATH = "../fonts/RictyDiminished-Regular.ttf"
    fm.fontManager.addfont(FONT_PATH)
    
    # FontPropertiesオブジェクト生成（名前の取得のため）
    font_prop = fm.FontProperties(fname=FONT_PATH)
    font_prop.get_name()
    
    plt.rcParams['font.family'] = font_prop.get_name()
    plt.rcParams['font.size'] = 16
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    mpf.plot(df, type='candle', style='default', datetime_format='%y-%m-%d', xrotation=0, ax=ax)
    ax.grid()

    return ax
```

```python
code = 'SOX'
symbol = '^%s' % code
ticker = yf.Ticker(symbol)
```

```python
# 2024-08-05
df1 = ticker.history(start='2024-07-22', end='2024-08-24', interval='1d')
df1
```

```python
ax1 = chart_candle(df1)
ax1.set_title('Daily chart for %s (%s)\n2024年8月5日の急落時' %(ticker.info['longName'], symbol))

plt.tight_layout()
#plt.savefig('report_crash_2024-08-05_%s.png' % code)
plt.show()
```

```python
ylim1 = ax1.get_ylim()
```

```python
# 2025-04-07
df2 = ticker.history(start='2025-03-24', end='2025-04-26', interval='1d')

for day_str in ['2025-04-21', '2025-04-22', '2025-04-23', '2025-04-24', '2025-04-25']:
    d = pd.to_datetime('%s 00:00:00-04:00' % day_str)
    df2.loc[d, 'Volume'] = 0

print(df2)
```

```python
ax2 = chart_candle(df2)
ax2.set_title('Daily chart for %s (%s)\n2025年4月2日夕方（米国時間）、トランプ関税発動' %(ticker.info['longName'], symbol))

plt.tight_layout()
#plt.savefig('report_crash_2025-04-07_%s.png' % code)
plt.show()
```

```python
ylim2 = ax2.get_ylim()
```

```python
y_min = np.min(ylim1 + ylim2)
y_max = np.max(ylim1 + ylim2)
print(y_min, y_max)
```

```python
ax1 = chart_candle(df1)
ax1.set_title('Daily chart for %s (%s)\n2024年8月5日の急落時' %(ticker.info['longName'], symbol))
ax1.set_ylim(y_min, y_max)

plt.tight_layout()
plt.savefig('report_crash_2024-08-05_%s.png' % code)
plt.show()
```
