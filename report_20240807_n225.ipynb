```python
tse = "https://www.jpx.co.jp/markets/statistics-equities/misc/tvdivq0000001vg2-att/data_j.xls"
filename = wget.download(tse)
df_tickers = pd.read_excel(filename)[['コード', '銘柄名', '市場・商品区分']]
os.remove(filename)
df_tickers
```

```python
sorted(list(set(df_tickers['市場・商品区分'])))
```

```python
list_market = [
    'ETF・ETN',
    'グロース（内国株式）',
    'スタンダード（内国株式）',
    'プライム（内国株式）',
]
df_targets = df_tickers[df_tickers['市場・商品区分'].isin(list_market)]
df_targets
```

```python
df_targets = pd.concat([df_targets['コード'].astype(str), df_targets['銘柄名'], df_targets['市場・商品区分']], axis=1)
```

```python
#start = '2024-08-02'
#end = '2024-08-07'
dict_close_1 = dict()
dict_close_2 = dict()
dict_close_3 = dict()
dict_recovery = dict()

for code in df_targets['コード']:
    code2 = '%s.T' % code
    ticker = yf.Ticker(code2)
    df = ticker.history(start=start, end=end)
    if len(df) != 3:
        continue

    price_close_1 = float(df['Close'].iloc[0])
    price_close_2 = float(df['Close'].iloc[1])
    drop = price_close_1 - price_close_2
    mean = (price_close_1 + price_close_2) / 2
    if drop / mean <= 0.05:
        continue

    price_close_3 = float(df['Close'].iloc[2])
    up = price_close_3 - price_close_2

    recovery = up / drop
    dict_close_1[code] = price_close_1
    dict_close_2[code] = price_close_2
    dict_close_3[code] = price_close_3
    dict_recovery[code] = recovery
```

```python
df_close_1 = pd.DataFrame(
    dict_close_1,
    index=['close1']
).T
df_close_2 = pd.DataFrame(
    dict_close_2,
    index=['close2']
).T
df_close_3 = pd.DataFrame(
    dict_close_3,
    index=['close3']
).T
df_recovery = pd.DataFrame(
    dict_recovery,
    index=['ratio']
).T

df_result = pd.concat([df_close_1, df_close_2, df_close_3, df_recovery], axis=1)
df_result
```

```python
df_result_top = df_result.sort_values('ratio', ascending=False).head(20)
```

```python
df_result_top_asc['cname'] = ['%s (%s)' % (df_targets.loc[df_targets['コード'] == code, '銘柄名'].iloc[0], code) for code in df_result_top_asc.index]
```
