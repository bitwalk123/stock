```python
import matplotlib.pyplot as plt
import mplfinance as mpf
import pandas as pd
import yfinance as yf

#symbol = '^IXIC'
symbol = '^GSPC'
ticker = yf.Ticker(symbol)
df = ticker.history(period='max', interval='1mo')
#df = ticker.history(start='1993-01-01', end='1998-01-01', interval='1mo')
#df = ticker.history(start='2019-11-01', end='2024-11-01', interval='1mo')
df
```

```python
ticker.info
```

```python
mswin = dict(
    vlines=[
        '1985-11-20', # Windows 1
        '1987-12-09', # Windows 2
        '1990-05-22', # Windows 3
        '1993-07-27', # Windows NT
        '1995-08-24', # Windows 95
        '1998-06-25', # Windows 98
        '1999-12-15', # Windows 2000
        '2000-09-14', # Windows Me
        '2001-08-24', # Windows XP
        '2006-11-30', # Windows Vista
        '2009-09-01', # Windows 7
        '2012-08-16', # Windows 8
        '2015-07-29', # Windows 10
        '2021-10-05', # Windows 11
    ],
    linewidths=0.5,
    colors=['C0', 'C0', 'C0', 'C0', 'C1', 'C0', 'C0', 'C0', 'C1', 'C0', 'C1', 'C0', 'C1', 'C1']
)
```

```python
fig = plt.figure(figsize=(10, 6))
#ax = fig.add_axes(111)
ax = fig.add_subplot(111)

mpf.plot(
    df, type='candle', style='yahoo',
    volume=False,
    vlines=mswin,
    datetime_format='%Y-%m-%d',
    xrotation=0,
    ax=ax,
)
ax.set_title('Monthly chart for %s' % ticker.info['longName'])
ax.grid()

plt.tight_layout()
plt.savefig('monthly_chart.png')
plt.show()
```
