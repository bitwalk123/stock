```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import mplfinance as mpf
import pandas as pd
```

```python
csvname = 'ohlc_heikin.csv'
df = pd.read_csv(csvname, index_col=0)
df.index = pd.to_datetime(df.index)
df
```

```python
plt.rcParams["font.size"] = 14
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)

mpf.plot(
    df,
    type='candle',
    style='binance',
    volume=False,
    datetime_format='%H:%M',
    xrotation=0,
    ax=ax,
)
ax.grid()

plt.tight_layout()
ax.set_title('ローソク足チャート (平均足)')

plt.savefig('../screenshots/ohlc_heikin.png')
plt.show()
```
