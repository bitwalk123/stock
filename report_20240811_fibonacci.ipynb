## Calculating Fibonacci retracement levels using Python

As we now know, retracements are the price movements that go against the original trend. To forecast the Fibonacci retracement level we should first identify the total up move or total down move. To mark the move, we need to pick the most recent high and low on the chart.

Letâ€™s take an example of Exxon Mobil to understand the Fibonacci retracement construction.

```python
# To import stock prices
import yfinance as yf


# To plot
import matplotlib.pyplot as plt


# Get stock data for XOM from Yahoo Finance
df = yf.download('XOM', start='2020-08-01', end='2024-04-08')


# Plotting
fig, ax = plt.subplots()
ax.plot(df['Close'], color='black')


# Adding labels to x and y axis
plt.xlabel('Date')
plt.ylabel('Exxon Mobil Corp (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Get the minimum and maximum price in the given date range
price_max = df['Close'].max()
price_min = df['Close'].min()

print("Minimum Price:", price_min)
print("Maximum Price:", price_max)
```

```python
import matplotlib.pyplot as plt

# Fibonacci Levels considering original trend as upward move
diff = price_max - price_min
level1 = price_max - 0.236 * diff
level2 = price_max - 0.382 * diff
level3 = price_max - 0.618 * diff

# Print Fibonacci levels
print("Level\tPrice")
print("0\t", price_max)
print("0.236\t", level1)
print("0.382\t", level2)
print("0.618\t", level3)
print("1\t", price_min)


# Set y-axis limits based on minimum and maximum price (reversed for ascending order)
plt.ylim(price_min - 0.05 * diff, price_max + 0.05 * diff)  # Expand the y-axis range slightly

# Plot close price
plt.plot(df.index, df['Close'], color='black', label='Close Price')

# Plot horizontal lines for each Fibonacci level with individual colors
plt.axhline(price_max, color='red', linestyle='--', label='Price Max')
plt.axhline(level1, color='green', linestyle='--', label='23.6%')
plt.axhline(level2, color='blue', linestyle='--', label='38.2%')
plt.axhline(level3, color='purple', linestyle='--', label='61.8%')
plt.axhline(price_min, color='orange', linestyle='--', label='Price Min')

# Add labels and legend
plt.ylabel("Price")
plt.xlabel("Date")
plt.legend()
plt.xticks(rotation=45)

# Show plot
plt.show()
```

It is visible that the maximum price is 121.37 where the level is 0 since there is no retracement at the maximum price.

The first retracement level is at 23.6% is $100.17, the second retracement level is at 38.2% is $87.06, and the next retracement level is at 61.8% is $65.87.
