```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import yfinance as yf

#ターゲットを指定
ticker = '^N225'

#データを収集
data = yf.download(ticker , period='12mo', interval = '1d')
```

```python
#pd.to_datetime(data['Date'], format='%Y-%m-%d')
data.to_csv('N225.csv')
```

```python
df = data
df = df.drop('Volume', axis=1)

#可視化する
df.plot()
```
