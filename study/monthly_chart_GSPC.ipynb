```python
import matplotlib.pyplot as plt
import mplfinance as mpf
import pandas as pd
import yfinance as yf

symbol = '^GSPC'
ticker = yf.Ticker(symbol)
#df = ticker.history(period='max', interval='1mo')
#df = ticker.history(start='1993-01-01', end='1998-01-01', interval='1mo')
df = ticker.history(start='2019-11-01', end='2024-11-01', interval='1mo')
df
```

```python
ticker.info
```

```python
fig = plt.figure(figsize=(10, 6))
#ax = fig.add_axes(111)
ax = fig.add_subplot(111)

mpf.plot(
    df, type='candle', style='yahoo',
    volume=False,
    datetime_format='%Y-%m-%d',
    xrotation=0,
    ax=ax,
)
ax.set_title('Monthly chart for %s' % ticker.info['longName'])
ax.grid()

plt.tight_layout()
plt.savefig('monthly_chart.png')
plt.show()
```
